**Teoría de Estabilidad de Sistemas de Ecuaciones Diferenciales Ordinarias Autónomas.**

Estudiaremos los problemas de valor inicial

$$\begin{equation*}
\frac{d \vec{x}}{d t} = A \vec{x}, \ \mathrm{con} \ \vec{x}(0)=\vec{x_0}
\end{equation*}$$
donde $A$ es una matriz cuadrada constante de dimensión $n \times n$

Si $n=1$, la solución de la ecuación diferencial es

$$\begin{equation*}
\vec{x}(t) = x_0 \exp{(At)}
\end{equation*}$$
Si $n = 1$, la solución también tendrá la misma forma, pero con $A$ siendo una matriz.

Podemos definir la exponencial de una matriz cuadrada constante A de dimensiones $n \times n$

$$\begin{equation*}
\exp(A) = \sum_{k=0}^{\infty} \frac{A^k}{k!}
\end{equation*}$$
Esto es similar al desarrollo en Serie de Taylor de $\exp(x)$, alrededor de $x_0=0$.

Vimos en el tema de Álgebra Lineal, que si la matriz es diagonalizable (todos sus eigenvalores son distintos), entonces existe una matriz $P$ formada por los eigenvectores de $A$, tal que $A=P^{-1} \Lambda P$, donde $\Lambda$ es la matriz diagonal formada por los $n$ eigenvalores $(\lambda_1, \lambda_2, \ldots, \lambda_n)$.

También vimos que el cálculo de potencias de la matriz $A$ se simplificaban computacionalmente

$$\begin{equation*}
A^k = P^{-1} \Lambda^k P
\end{equation*}$$
Si podemos diagonalizar la matriz $A$ ($n$ eigenvalores distintos), y $P = [ x_1 x_2  \ldots  x_n ]$ es un eigenvector de $A$, tendremos

$$\begin{equation*}
A = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \lambda_n \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]   
\end{equation*}$$
por lo que entonces se tiene que

$$\begin{equation*}
At = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\lambda_1 t & 0 & \cdots & 0 \\
0 & \lambda_2 t & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \lambda_n t \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]   
\end{equation*}$$
y calculando la exponencial de $\exp(At)$ sería equivalente a calcular

$$\begin{equation*}
\exp(At) = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\exp(\lambda_1 t) & 0 & \cdots & 0 \\
0 & \exp(\lambda_2 t) & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \exp(\lambda_n t) \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]  
\end{equation*}$$
Es fácil ver el comportamento de la solución cuando $\lim t \rightarrow  \infty$, lo determinan las exponenciales de los eigenvalores.

**Sistema de ecuaciones lineales homogéneas.**

Supongamos que tenemos el siguiente sistema de ecuaciones diferenciales

$$\begin{eqnarray*}
\frac{d x_1}{dt} & = & a x_1 + b x_2 \\
\frac{d x_2}{dt} & = & c x_1 + d x_2 \\
\end{eqnarray*}$$
Si $\vec{x}^T = (x_1, x_2)$, entonces la ecuación anterior se puede escribir como

$$\begin{equation*}
\frac{d \vec{x}}{dt} = \frac{d}{dt} \begin{bmatrix}
x_1 \\ x_2 \end{bmatrix} = A \begin{bmatrix}
x_1 \\ x_2 \end{bmatrix}, \mathrm{donde} \ A = \begin{bmatrix}
a & b \\ c & d \end{bmatrix}
\end{equation*}$$

Si definimos como $p = \mathrm{Tr}\ (A) = (a+d)$ y $q=\det (A) = (ad-bc)$

El polinomio característico de $\det(A - \lambda I)$ satisface

$$\begin{equation*}
p(\lambda) = \lambda^2 - p \lambda + q  
\end{equation*}$$
cuyas raíces son

$$\begin{equation*}
\lambda_{1,2} = \frac{p \pm \sqrt{p^2 - 4 q}}{2}
\end{equation*}$$

donde el discriminante $\Delta = p^2 - 4q$ puede ser mayor o menor que cero (es decir, $\lambda$ Real o Compleja).

Se puede ver que $p$ y $q$ satisfacen:

$$\begin{eqnarray*}
p & = & \lambda_1 + \lambda_2 \\
q & = & \lambda_1 \cdot \lambda_2 \\
\end{eqnarray*}$$

**Clasificación de las soluciones**

Para sistema de matrices 2x2, podemos tener 3 posibles categorías:

$$\begin{eqnarray*}
A) \begin{bmatrix}
\lambda & 0 \\
0 & \mu
\end{bmatrix}; B) \begin{bmatrix}
\lambda & 0 \\
1 & \lambda
\end{bmatrix}; C) \begin{bmatrix}
a & -b \\
b & a
\end{bmatrix}, b\neq0
\end{eqnarray*}$$

Estas 3 categorías pueden dividirse en 14 casos de puntos críticos:

**Caso A)** ($\Delta > 0$, eigenvalores reales y distintos)
$$\begin{equation*}
 \begin{bmatrix}
\lambda & 0 \\
0 & \mu
\end{bmatrix}
\end{equation*}$$

La solución se puede escribir como

$$\begin{equation*}
x(t) = c_1 \exp(\lambda t) \ {v_1} + c_2 \exp(\mu t) \ {v_2}
\end{equation*}$$
donde ${v_1}$ y ${v_2}$ son los eigenvectores correspondentes a los eigenvalores $\lambda$ y $\mu$ respectivamente, $c_1$ y $c_2$ constantes.

Puntos críticos:

- Punto silla: $\lambda < 0 < \mu$
- Nodo estable: $\lambda < \mu < 0$
- Nodo estable: $\lambda = \mu < 0$
- Nodo inestable: $0 < \lambda < \mu$
- Nodo inestable: $0 < \lambda = \mu$
- Nodo degenerado: $\lambda < \mu = 0 $
- Nodo degenerado: $0 = \mu < \lambda$
- Nodo degenerado: $0 = \mu = \lambda$

**Caso B)** ($\Delta = 0$, eigenvalores reales e iguales)
$$\begin{equation*}
 \begin{bmatrix}
\lambda & 0 \\
1 & \lambda
\end{bmatrix}
\end{equation*}$$

La solución se puede escribir como

$$\begin{equation*}
x(t) = [ (c_1 + c_2 t) \ v_1 + c_2 v_2 ] \ \exp(\lambda t)
\end{equation*}$$
donde $c_1$ y $c_2$ son constantes y los vectores $v_1$ y $v_2$ satisfacen las ecuaciones

$$\begin{eqnarray*}
A v_1 & = & \lambda v_1 \\
A v_2 & = & \lambda v_1 + v_2
\end{eqnarray*}$$

Puntos críticos:

- Nodo estable: $\lambda < 0$
- Nodo inestable: $0 < \lambda$
- Nodo degenerado: $\lambda = 0$

**Caso C)**($\Delta < 0$, eigenvalores complejos)
$$\begin{equation*}
 \begin{bmatrix}
a & -b \\
b & a
\end{bmatrix}, b \neq 0
\end{equation*}$$

El polinomio característico es

$$\begin{eqnarray*}
(a-\lambda)^2 + b^2 & = & 0 \\
\lambda^2 + a^2 - 2 a \lambda + b^2 & = & 0 \\
\lambda^2 - 2 a \lambda + (a^2+b^2) & = & 0 \\  
\end{eqnarray*}$$
cuyas raíces son complejas:

$$\begin{equation*}
\lambda_{1,2} = a \pm i b
\end{equation*}$$
La solución se puede escribir como

$$\begin{eqnarray*}
x(t) & = & c_1 \exp((a+ib)t) \ \ v_1 + c_2 \exp((a-ib)t) \ v_2 \\
& = & c_1 \exp(at) \exp(ibt) \ v_1 +
c_2 \exp(at) \exp(-ibt) \ v_2 \\
& = & [c_1 \exp(ibt) \ v_1 +
c_2 \exp(-ibt) \ v_2] \ \exp(at) \\
& = & [c_1 (\cos(bt) + i \sin(bt)) \ v_1 +
c_2 (\cos(bt) - i \sin(bt)) \ v_2 ] \exp(at) \\
\end{eqnarray*}$$

donde $c_1$ y $c_2$ son constantes complejas y los vectores $v_1$ y $v_2$ son vectores complejos en $C^2$.

Puntos críticos:

Foco o Espiral estable: $a < 0 < b \ $ ($\mathrm{Tr}(A) < 0$)

Foco o Espiral inestable: $b < 0 < a \ $ ($\mathrm{Tr}(A) > 0$)

Centro: $a=0, b>0 \ $ ($\mathrm{Tr}(A) = 0$)

**Equilibrio de Sistemas No Lineales**

Estudiaremos sistemas de ecuaciones diferenciales no lineales cuyos puntos críticos son el origen.

Veremos que las soluciones de los sistemas no lineales cerca del origen se comportan como las soluciones de sistemas linealizados alrededor del origen.

Analicemos el siguiente sistema de ecuaciones diferenciales no lineales

$$\begin{eqnarray}
\frac{dx}{dt} & = & f(x,y) \\
\frac{dy}{dt} & = & g(x,y) \\
\end{eqnarray}$$

Supongamos que el punto crítico es ($x,y$)=($\bar{x},\bar{y}$), donde se anula el lado derecho del sistema de ecuaciones, es decir $f(\bar{x}, \bar{y}) = 0$ y $g(\bar{x}, \bar{y}) = 0$.

Nos interesa saber si el punto crítico es estable o inestable.

Consideremos una pequeña perturbación alrededor del punto crítico

$$\begin{eqnarray*}
x & = & \bar{x} + u  \\
y & = & \bar{y} + v  \\
\end{eqnarray*}$$
donde $u$ y $v$ son pequeñas. La pregunta de estabilidad del punto crítico se replantea de si $u$ y $v$ crecerán infinitamente o no.

Podemos escribir el sistema de ecuaciones que satisfacen $u$ y $v$

$$\begin{eqnarray*}
\frac{du}{dt} & = & \frac{dx}{dt} \\
              & = & f(x,y) \\
              & = & f(\bar{x} + u, \bar{y} + v) \\
              & = & f(\bar{x}, \bar{y}) + 
              \frac{\partial f}{\partial x}(\bar{x}, \bar{y}) \ u + \frac{\partial f}{\partial y}(\bar{x}, \bar{y}) \ v + \cdots \\
              & = & \frac{\partial f}{\partial x}(\bar{x}, \bar{y}) \ u + \frac{\partial f}{\partial y}(\bar{x}, \bar{y}) \ v + \cdots \\
\end{eqnarray*}$$

De forma similar obtenemos la ecuación para $v$

$$\begin{eqnarray*}
\frac{dv}{dt} & = & \frac{\partial g}{\partial x}(\bar{x}, \bar{y}) \ u + \frac{\partial g}{\partial y}(\bar{x}, \bar{y}) \ v + \cdots \\
\end{eqnarray*}$$

Podemos suponer que los terminos $u^2, v^2, uv, \ldots$, son muy pequeños, por lo que podemos aproximar el sistema de ecuaciones por

$$\begin{equation*}
\frac{d}{dt}\begin{bmatrix} 
u \\ 
v \\ 
\end{bmatrix} = \begin{bmatrix}
\frac{\partial f}{\partial x}(\bar{x}, \bar{y}) & \frac{\partial f}{\partial y}(\bar{x}, \bar{y}) \\
\frac{\partial g}{\partial x}(\bar{x}, \bar{y})  & \frac{\partial g}{\partial y}(\bar{x}, \bar{y}) \\
\end{bmatrix}
\begin{bmatrix}
u \\
v \\
\end{bmatrix} = {\bf J} \begin{bmatrix}
u \\
v \\
\end{bmatrix}
\end{equation*}$$

La matriz ${\bf J}$ es la matriz Jacobiana, evaluada el el punto crítico $(\bar{x}, \bar{y})$.

El sistema anterior tiene el punto crítico $(\bar{x}, \bar{y}) = (0,0)$, y los eigenvalores de la matriz ${\bf J}$ determinan la naturaleza de los puntos críticos, como los que se vió en este tema.

En este caso podemos afirmar en sólo dos casos:

- Si la parte real de los eigenvalores son positivos, el punto crítico es inestable.
- El caso contrario, si la parte real de los eigenvalores son negativos, entonces el punto crítico es estable.

Cualquier otro resultado (por ej. parte real cero), no podremos decir mucho, sino que requeriremos incluir más términos de orden superior, ya que estaremos en una frontera no bien determinada.

Un resultado del Teorema de Hartman-Grobman garantiza que la estabilidad del punto crítico del sistema original $(\bar{x}, \bar{y})$ es la misma que el punto crítico $(0,0)$ del sistema linearizado.

**Ejemplo: Sistema de Lotka Volterra**

El sistema de ecuaciones de Lotka-Volterra, describe la dinámica poblacional de dos especies: presa-depredador, por ejemplo conejos y zorros.

El sistema de ecuaciones que describe la variación de la población, está dado por

$$\begin{eqnarray*}
\frac{dx}{dt} & = &  a x - b x y \\ 
\frac{dy}{dt} & = &  -c y + d x y  \\
\end{eqnarray*}$$

donde $x$ es el número de presas (p. ej. conejos) y $y$ el número de depredadores (p. ej. zorros).

Los parámetros $b, d$ corresponden a las interacciones entre las especies, $a$ es la taza de natalidad de los conejos y $c$ la taza de extinción de los zorros si no hay alimento.

```
# Ejemplo de Lotka-Volterra

import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y

# Definimos los parámetros 
a = 1.
b = 0.1
c = 1.5
d = 0.75

def dX_dt(X, t=0):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([ a*X[0] -   b*X[0]*X[1] ,
                  -c*X[1] + d*b*X[0]*X[1] ])
```

```

# Los puntos críticos del sistema
X_f0 = np.array([     0. ,  0.])
X_f1 = np.array([ c/(d*b), a/b])

print('Puntos críticos del sistema para estos valores de parámetros')
print('X_f0:', X_f0)
print('X_f1:', X_f1)

# En los puntos críticos se anulan las derivadas. 
all(dX_dt(X_f0) == np.zeros(2) ) and all(dX_dt(X_f1) == np.zeros(2))

def d2X_dt2(X, t=0):
    """ El Jacobiano evaluado en X}. """
    return np.array([[a -b*X[1],   -b*X[0]     ],
                  [b*d*X[1] ,   -c +b*d*X[0]] ])
```

```
# El Jacobiano evaluado en el primer punto crítico X_f0
A_f0 = d2X_dt2(X_f0)
print('A_f0:')
print(A_f0)
```

```
# El Jacobiano evaluado en el segundo punto crítico X_f1
A_f1 = d2X_dt2(X_f1)                    
print('A_f1:')
print(A_f1)
                                
# Y sus eigenvalores son +/- sqrt(c*a).j:
lambda1, lambda2 = linalg.eigvals(A_f1) 
print('Eigenvalores:', lambda1, lambda2)
# >>> (1.22474j, -1.22474j)

# Los eigenvalores son imaginarios
# Con periodo
T_f1 = 2*np.pi/abs(lambda1) 
print('Periodo:', T_f1)
```

```
# Integramos el sistema de ecuaciones de Lotka-Volterra usando odeint

t = np.linspace(0, 15,  1000)              # Rango de tiempo de integración
X0 = np.array([10, 5])                     # Condiciones iniciales: 10 conejos y 5 zorros
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']
```

```
# Graficamos las soluciones

conejos, zorros = X.T


f1 = plt.figure();
plt.figure(figsize=(10, 5)); 
plt.plot(t, conejos, 'r-', label='Conejos');
plt.plot(t, zorros  , 'b-', label='Zorros');
plt.grid();
plt.legend(loc='best');
plt.xlabel('tiempo');
plt.ylabel('población');
plt.title('Evolución de la población de conejos y zorros');
plt.show();
```

```
# Graficamos la familia de soluciones en el espacio fase para distintas condiciones iniciales

values  = np.linspace(0.3, 0.9, 5)                         # posición de X0 entre X_f0 y X_f1
vcolors = plt.cm.cividis(np.linspace(0.3, 1., len(values)))  # colores para cada trayectoria

plt.figure(figsize=(12, 8)); 

#-------------------------------------------------------
# grafica las trayectorias
for v, col in zip(values, vcolors):
    X0 = v * X_f1                               # punto inicial
    X = integrate.odeint( dX_dt, X0, t)         # Integración para cada condición inicial
    plt.plot( X[:,0], X[:,1], lw=3.5*v, color=col, label='X0=(%.f, %.f)' % ( X0[0], X0[1]) )

#-------------------------------------------------------
# define una malla y calcula la pendiente (dirección) en cada punto
ymax = plt.ylim(ymin=0)[1]                        # get axis limits
xmax = plt.xlim(xmin=0)[1]
nb_points   = 20

x = np.linspace(0, xmax, nb_points)
y = np.linspace(0, ymax, nb_points)

X1 , Y1  = np.meshgrid(x, y)                       # crea la malla
DX1, DY1 = dX_dt([X1, Y1])                      # calcula la tasa de crecimiento
M = (np.hypot(DX1, DY1))                           # distancia euclideana 
M[ M == 0] = 1.                                 # Evita divisiones entre 0 
DX1 /= M                                        # Normaliza el tamaño de cada flecha
DY1 /= M

#-------------------------------------------------------
# Dibujar los campos de direcciones usando la función quiver de Matplotlib 
# EL color de las flechas indica la rapidez de crecimiento

plt.title('Plano Fase: Trayectorias y campo de direcciones - Punto crítico (20,10)')
Q = plt.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=p.cm.jet)
plt.xlabel('Número de conejos')
plt.ylabel('Número de zorros')
plt.legend(loc='best')
plt.grid()
plt.xlim(0, xmax)
plt.ylim(0, ymax)
plt.show()
```

Solución Ejemplo 7:

$$\begin{eqnarray*}
\frac{dx}{dt} & = & 2x+y \\
\frac{dy}{dt} & = & x+y \\
\end{eqnarray*}$$

Con condiciones iniciales: $x(1)=1$ y $y(1)=1$.

Equivale al problema

$$\begin{equation*}
\frac{d}{dt}\begin{bmatrix} x \\ y \end{bmatrix} = A \begin{bmatrix} x \\ y \end{bmatrix}
\end{equation*}$$

Donde la matriz $A$ está definida por

$$\begin{bmatrix}
2 & 1 \\
1 & 1 \\
\end{bmatrix}$$
Definimos $X = [x,y]^T$.

La solución de este sistema es

$$\begin{equation*}
X(t) = X(1) \exp(At) = X(1) P \exp(\Lambda t) P^{-1}
\end{equation*}$$

Donde $P$ es la matriz de eigenvectores y $\Lambda$ la matriz de los eigenvalores $\lambda_1$ y $\lambda_2$ de la matriz de transformación $A$ (det$A \neq 0$)

```
# Resolviendo Ejercicio 7

import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y
# Definimos la Matriz A 

A7 = np.array([[2,1], [1,1]])
print('A =', A7)

def dX_dt(X, t=1):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([ 2*X[0] + X[1] ,
                  X[0]+X[1]])
    
# Los puntos críticos del sistema
X_f0 = np.array([     0. ,  0.])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
all(dX_dt(X_f0) == np.zeros(2) )
                                
# usamos scipy.linalg.eig()
eigvals, eigvecs = linalg.eig(A7)
print('Eigenvalores:', eigvals)
print('Eigenvectores:', eigvecs)
```

**Ejemplo de Solución de un sistema 3x3**

Resolveremos el sistema

$$\begin{eqnarray*}
x_1' & = & x_1 \\
x_2' & = & x_1 + 2 x_2 \\
x_3' & = & x_1 - x_3 \\
\end{eqnarray*}$$
con condiciones iniciales $\vec{X}(0)=(1,0,0)$.

O si lo escribimos de forma matricial

$$\begin{equation*}
\vec{X'} = \begin{bmatrix} 1 & 0 & 0 \\
1 & 2 & 0 \\
1 &  0 & -1\\ \end{bmatrix} \vec{X} =
A \vec{X}
\end{equation*}$$
Buscaremos los eigenvalores y eigenvectores de la matriz $A$

$$\begin{equation*}
A = P \Lambda P^{-1}
\end{equation*}$$
donde $P$ es la matriz de los eigenvectores y $\Lambda$ la matriz diagonal de los eigenvalores.

```

import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y, X[2]=z
# Definimos la Matriz A 

A3 = np.array([[1,0,0],[1,2,0], [1,0,-1]])
print('A =', A3)

def dX_dt(X,t=0):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([X[0], X[0] + 2*X[1] ,
                  X[0]-X[1]])
    
# Los puntos críticos del sistema
X_f0 = np.array([ 0.0 , 0.0, 0.0])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
# En los puntos críticos se anulan las derivadas. 
all(dX_dt(X_f0) == np.zeros(3) )
                                
# Aplicamos la biblioteca scipy.linalg.eig()
eigvals, eigvecs = linalg.eig(A3)
print('Eigenvalores:', eigvals)
print('Eigenvectores:', eigvecs)
```

```
# Solución
# Los eigenvectores son linealmente independientes
lambda1 = eigvals[0]
lambda2 = eigvals[1]
lambda3 = eigvals[2]
print('Eigenvalores')
print('lambda1= ', lambda1)
print('lambda2= ', lambda2)
print('lambda3= ', lambda3)
print(' ')
print('Eigenvectores')
v1 = eigvecs[:,0]
v2 = eigvecs[:,1]
v3 = eigvecs[:,2]
print('v1=',v1)
print('v2=',v2)
print('v3=',v3)
```

```
# La solución del sistema 
# Solución numérica
# Integramos el sistema de ecuaciones de Lotka-Volterra usando odeint

t = np.linspace(0, 3,  300)              # Rango de tiempo de integración
X0 = np.array([1, 0, 0])                     # Condiciones iniciales
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']
```

```
x, y, z = X.T

f1 = plt.figure();
plt.figure(figsize=(10, 5)); 
plt.plot(t, x, 'r-', label='x(t)');
plt.plot(t, y  , 'b-', label='y(t)');
plt.plot(t, z  , 'g-', label='z(t)');
plt.grid();
plt.ylim(-10, 10)
plt.legend(loc='best');
plt.xlabel('tiempo');
plt.ylabel('Soluciones');
plt.title('Soluciones x(t), y(t), z(t)');
plt.show();
```
